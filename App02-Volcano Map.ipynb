```python
#--------------------------------------------
# Adding points to the base map
#--------------------------------------------
TILES = "Stamen Terrain"
map = folium.Map(location=[39.97199, -83.00275], zoom_start=10.5, tiles=TILES)
map
```

```python
#--------------------------------------------
#  Save map image
#--------------------------------------------
map.save(image_file)
```

```python
#--------------------------------------------
#  Add points with CircleMarker feature
#--------------------------------------------
map.add_child(folium.Marker(location=[39.92, -83.02], popup="Hi Marker", icon=folium.Icon(color='green')))
map
```

```python
#--------------------------------------------------
#  Adding children to a map with a feature group
#--------------------------------------------------
fg = folium.FeatureGroup(name="My Map")
fg.add_child(folium.Marker(location=[40.105499578, -83.13749], popup="Dublin OH", icon=folium.Icon(color='green')))
map.add_child(fg)
map
```

```python
#------------------------------------------
#  Multiple Markers on Map
#------------------------------------------
map = folium.Map(location=[39.97199, -83.00275], zoom_start=10.5, tiles=TILES)
fg = folium.FeatureGroup(name="My Map")

for coordinates in [[40.105499578, -83.13749], [39.97199, -83.00275]]:

    fg.add_child(folium.Marker(location=coordinates, popup="Dublin OH", icon=folium.Icon(color='red')))
map.add_child(fg)
map
```
