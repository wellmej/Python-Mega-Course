```python
df=pd.read_csv(csv_file)
df
```

```python
df["Full Address"]=df["Address"] + ", " + df["City"] + ", " + df["State"] + ", " + df["Country"]
df
```

```python
df["Coordinates"]=df["Full Address"].apply(nom.geocode)
df
```

```python
df.Coordinates[0]
```

```python
df.Coordinates[0].latitude
```

```python
df.Coordinates[0].longitude
```

```python
#-------------------------------------------------------------------------------
#  Need to appy lambda function to set latitude and longitude columns
#-------------------------------------------------------------------------------
df["Latitude"]=df["Coordinates"].apply(lambda x: x.latitude if x != None else None)
df["Longitude"]=df["Coordinates"].apply(lambda x: x.longitude if x != None else None)
df
```

```python
#-------------------------------------------------------------------------------
#  Need to appy lambda function to set latitude and longitude columns
#  Note: Without Lambda function you get this error
#-------------------------------------------------------------------------------
df["Latitude"]=df["Coordinates"].latitude
df
```
